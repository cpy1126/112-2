# 作業2：Json資料處理
## 調查東南亞各國外國交換生就讀情形
## 最終分析結果:
# 1.哪個國家有最多的人出國讀書
# 2.哪個國家有最多的外國交換生
# 3.哪個國家有最多的大專附設華語中心的學生

資料來源：政府資料開放平台[（International Air Transport Association )](https://data.gov.tw/)

```python

import json
f=open('data.json',encoding="utf-8")

data=json.load(f)
print(data)
```

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt

# 假设您已经有了正确格式的数据
# 用实际数据替换这里的样本数据


# 将数据写入 CSV 文件
with open('new.csv', mode='w', newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(['國別', '學位生_正式修讀學位外國生'])
    for i in range(len(data["member"])):
        row = [data["member"][i]['國別'], data["member"][i]['學位生_正式修讀學位外國生']]
        writer.writerow(row)

# 使用 pandas 读取 CSV 文件中的数据
data = pd.read_csv('new.csv', encoding="utf-8")

# 绘制图表
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']
plt.xlabel("國別")
plt.ylabel("學位生_正式修讀學位外國生")
plt.title("哪個國家有最多的人出國讀書")
plt.bar(data['國別'], data['學位生_正式修讀學位外國生'])
plt.xticks(rotation=45)  # 旋转 x 标签以便更好地阅读
plt.show()

```
