```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
import re
import json
```

```python
# 目標網址
url = "https://www.vscinemas.com.tw/vsweb/film/index.aspx"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
soup
```

```python
# 找到所有class為'infoArea'的section標籤
info_areas = soup.find_all('section', class_='infoArea')

data1= []


# 迭代輸出每個section標籤中的h3標籤文字內容
for info_area in info_areas:
    title = info_area.find('h2')
    if title:
        print(title.text)
    subtitle = info_area.find('h3')
    if subtitle:
        print(subtitle.text)
    time = info_area.find('time')  
    if time:
        print(time.text)
        
    data1.append([title.text,subtitle.text,time.text])


with open('movie_info.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)

    # 寫入標題行
    writer.writerow(['Title', 'Subtitle','Time'])

    # 寫入資料行
    writer.writerows(data1)

print("CSV文件已生成：movie_info.csv")
    
```

```python
data = pd.read_csv('movie_info.csv')

# 顯示DataFrame
print(data)
```

```python
data2=[]

for info_area in info_areas:
    title = info_area.find('h2')
    if title:
        title_text=title.text
    subtitle = info_area.find('h3')
    if subtitle:
        subtitle_text=subtitle.text
    time = info_area.find('time')  
    if time:
        time_text=time.text
        data2.append({'title': title_text, 'subtitle': subtitle_text,'time':time_text})

df = pd.DataFrame(data2)
json_data = df.to_json(orient='records')

with open('movie_info.json', 'w', encoding='utf-8') as jsonfile:
    jsonfile.write(json_data)   

print("JSON檔案已生成：movie_info.json")
```
